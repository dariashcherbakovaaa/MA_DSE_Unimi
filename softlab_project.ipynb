```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/softlab-lab_finalproject/ESG_WorldBank_20042023_AllCountries_original.csv')
df.head(7)
```

```python
melted_df = pd.melt(df, id_vars=['Country Name', 'Country Code', 'Series Name', 'Series Code'], var_name='Year', value_name='Value')

# Extract the year from the column name
melted_df['Year'] = melted_df['Year'].str.extract(r'(\d{4})')

# Pivot the DataFrame
esg = melted_df.pivot_table(index=['Country Name', 'Country Code', 'Year'], columns='Series Name', values='Value', aggfunc='first').reset_index()

esg.head()
```

```python
metadf = pd.read_excel('/content/drive/MyDrive/softlab-lab_finalproject/Metadata.xlsx', sheet_name= "Series - Metadata")
print(f'The shape is {metadf.shape}\n\n')
metadf.head(6)
```

```python
metadf['Topic'] = metadf['Topic'].fillna('Environment')
metadf[metadf['Topic'].isna()].value_counts()
```

```python
import re

pattern_environment = r'\benvironment:\w*'


meta_env = metadf[metadf['Topic'].str.contains(pattern_environment, flags=re.IGNORECASE)]

print(f'The shape is {meta_env.shape}\n\n')
meta_env.head(6)
```

```python
meta_env.Topic.unique()
```

```python
print(meta_env['Indicator Name'].value_counts())
indicators_list = meta_env['Indicator Name'].unique()
```

```python
esg.columns.values.tolist()
```

```python
relevant_columns = ["Country Name", "Year"] + [col for col in esg.columns if col in indicators_list]
esg_env = esg[relevant_columns]
```
